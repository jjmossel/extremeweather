```python
stations_all = stations_info()
```

```python
stations = stations_all.loc[lambda df:df['state'].notnull() & (df['date_min']<'1960-01-01') & (df['date_max']>'2022-01-01')  & (df['date_count']>20000)]
stations_NY = stations.loc[lambda df:df['state']=='NY']
```

```python
PRCP_NY = hist_daily(stations_NY.index.values[0:4],'PRCP')
PRCP_NY.loc['1920-01-01':]
```
